## **Penetración de Internet fijo (accesos por cada 100 hogares)**
### **Número de accesos al servicio de Internet fijo por cada 100 hogares por provincia**
Es la relación proporcional de los accesos del servicio de Internet fijo por las diferentes tecnologías de acceso (cable coaxial, DSL, satélite, terrestre fijo inalámbrico, y otras tecnologías) respecto al total. Método de cálculo: Para el cálculo de la penetración se considera número de hogares.

### **Internet penetración por periodo**

```python
InternetPenetracion_df = pd.read_excel("Data_ETL/InternetPenetracion.xlsx")
InternetPenetracion_df = InternetPenetracion_df.sort_values(by=['Año','Trimestre'], ascending=True)
InternetPenetracion_df.head(5)
```

```python
# Internet penetración por cada 100 hogares
plt.figure(figsize=(12,5))
sns.set(style="darkgrid")
sns.lineplot(data=InternetPenetracion_df, x='Periodo', y='Accesos por cada 100 hogares',color='aqua', linewidth = 3)
plt.title('Internet Penetración')
plt.xticks(rotation=90);
```

```python
print("Minimo de accesos: ",InternetPenetracion_df['Accesos por cada 100 hogares'].min())
print("Máximo de accesos: " ,InternetPenetracion_df['Accesos por cada 100 hogares'].max())
```

La penetración de internet ha incrementado de un 50 cada 100 horares tenian acceso a un 76 cada 100 horares con acceso

```python
# Internet penetración por cada 100 habitates
plt.figure(figsize=(12,5))
sns.set(style="darkgrid")
sns.lineplot(data=InternetPenetracion_df, x='Periodo', y='Accesos por cada 100 hab',color='darkcyan', linewidth = 3)
plt.title('Internet Penetración')
plt.xticks(rotation=90);
```

```python
print("Minimo de accesos: ",InternetPenetracion_df['Accesos por cada 100 hab'].min())
print("Máximo de accesos: " ,InternetPenetracion_df['Accesos por cada 100 hab'].max())
```

Se pasó de 15 cada 100 personas con acceso a internet en el periodo de Enero-Marzo 2014 a 23 cada 100 en el periodo de Julio-Septiembre 2022

### **Por periodo**

```python
VelocidadMediaDeBajadaDeInternetNacional_df = pd.read_excel("Data_ETL/VelocidadMediaDeBajadaDeInternetNacional.xlsx")
VelocidadMediaDeBajadaDeInternetNacional_df  = VelocidadMediaDeBajadaDeInternetNacional_df .sort_values(by=['Año','Trimestre'], ascending=True)
VelocidadMediaDeBajadaDeInternetNacional_df.head()
```

```python
plt.figure(figsize=(12,5))
sns.set(style="darkgrid")
sns.lineplot(data=VelocidadMediaDeBajadaDeInternetNacional_df, x='Periodo', y='Mbps (Media de bajada)',color='BlueViolet')
plt.title('Velocidad Media de Bajada de Internet Fijo')
plt.xticks(rotation=90);
```

*Podemos ver que la velocidad media de bajada incrementa con tendencia polinomica cuadratica por tirmestre*

### **Por Provincia**

```python
HistoricoVelocidadInternet_df = pd.read_excel("Data_ETL/HistoricoVelocidadInternet.xlsx")
HistoricoVelocidadInternet_df.drop_duplicates(inplace=True)
HistoricoVelocidadInternet_df = HistoricoVelocidadInternet_df.reset_index(drop=True)
HistoricoVelocidadInternet_df.head(5)
```

```python
HistoricoVelocidadInternet_df.info()
```

```python
lista_provincias =list(HistoricoVelocidadInternet_df['Provincia'].unique())
lista_provincias_df =list(HistoricoVelocidadInternet_df['Provincia'].unique())
i = 0
for prov in lista_provincias:
    lista_provincias_df[i] = HistoricoVelocidadInternet_df[HistoricoVelocidadInternet_df['Provincia']==prov]
    i = i + 1 
```
